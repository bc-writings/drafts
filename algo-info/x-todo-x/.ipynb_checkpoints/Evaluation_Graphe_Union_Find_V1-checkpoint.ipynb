```python
G1_dico = {
    0 : [1,2,3,4],
    1 : [5, 6],
    2 : [3],
    4 : [9, 8],
    5 : [6],
    7 : [5, 6],
    8 : [9],
    10: [9, 8],
}
```

```python
max_sommet = -1

for sommet, voisins in G1_dico.items():
    max_sommet = max(sommet, *voisins, max_sommet)

G1 = [set() for _ in range(max_sommet + 1)]

for sommet, voisins in G1_dico.items():
    G1[sommet] |= set(voisins)
    
    for vs in voisins:
        G1[vs].add(sommet)
        
for sommet, voisins in enumerate(G1):
    G1[sommet] = sorted(list(voisins))
    
print(G1)
```

```python
def degre(G):
    return [len(voisins) for voisins in G]
```

```python
# [4, 3, 2, 2, 3, 3, 3, 2, 3, 3, 2] ?
print(degre(G1)) 
```

```python
#Fonction qui détermine et classe les degrés des sommets dans un ordre décroissant

def trieSommets(G):
    """
    Entrée : Une liste d'ajacence 
    Sortie : Liste de ses sommets trié par orde décroissant de degré
    
    """ 
    deg_2_sommets = {}
    
    for sommet, deg in enumerate(degre(G)):
        if deg not in deg_2_sommets:
            deg_2_sommets[deg] = []
            
        deg_2_sommets[deg].append(sommet)
        
    
    sommets = []
    
    for deg in sorted(deg_2_sommets.keys(), key = lambda x: -x):
        sommets += deg_2_sommets[deg]
    
    return sommets
```

```python
def reunir(P1, P2):
    reunion = list(set(P1 + P2))
    reunion.sort()

    return reunion       
```

```python
#Création d'une fonction où je recherche les sommets non adjacents au sommet traité 
def estNonAjacent(M,sommet):
    """
    Entrée : Une liste d'ajacence et un sommet
    Sortie : Revoie une liste contenant les sommets non adjacents
    """
    liste_non_ajacents=[]
    for i in range(len(M)):
        liste_non_ajacents.append(i)
    #print(liste_non_ajacents)
       
    for j in range(len(M[sommet])):
        for i in range(len(liste_non_ajacents)):
            #print(M[sommet][j])
            if M[sommet][j] in liste_non_ajacents:
                liste_non_ajacents.remove(M[sommet][j])
                #print(liste_non_ajacents)
    liste_non_ajacents.remove(sommet)
    return liste_non_ajacents
```

```python
#Fonction qui détermine et classe les degrés des sommets dans un ordre décroissant


def trieSommets_M(G):
    """
    Entrée : Une matrice adjacente
    Sortie : Liste de ses sommets trié par orde décroissant
    
    """
    listeSommetsDecroissant=[]
    liste=degre_M(G)
 
    #Recherche du maximum dans la liste des degrés
    maximum=liste[0]
   
    for i in range (len(liste)):
        if liste[i]>=maximum:
            maximum=liste[i]
            indice=i
    listeSommetsDecroissant.append(indice)
       
    
    #Recherche du minimum dans la liste des degrés
    minimum=liste[0]
    for i in range (len(liste)):
        if liste[i]<minimum:
            minimum=liste[i]
        
    j=maximum-1
    while j>=minimum:
        for k in range(len(liste)):
            if liste[k]==j:
                listeSommetsDecroissant.append(k)
        j-=1
        
    return listeSommetsDecroissant
     

trieSommets_M(liste_M)        
#trieSommets_M(liste_M1)
```

```python
#Création d'une fonction où je recherche les sommets non adjacents au sommet traité 
def estNonAjacent_M(M,sommet):
    """
    Entrée : Une martrice M
    Sortie : Revoie une liste contenant les sommets non adjacent si l'élément s1 ne sont pas adjacents
    """
    liste_non_ajacents=[]
    for i in range (sommet+1,len(M[0])):
        if M[sommet][i]==0:
            liste_non_ajacents.append(i)
    return liste_non_ajacents

estNonAjacent_M(liste_M,4)
```
