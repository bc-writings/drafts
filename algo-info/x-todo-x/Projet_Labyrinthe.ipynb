```python
# avec les forets
def racine(F,s):
    if F[s]==s:
        return s
    else:
        r=racine(F,F[s])  # optimisation, les hauteurs sont plus petites
        F[s]=r
    return r

def reunir(F,s1,s2):
    F[racine(F,s1)]=racine(F,s2)
    
# composantes connexes
def compConnexes(G): # G est la liste des listes des voisins de chaque sommet :
    G[s] = liste des voisins de s
    N = len(G) # nb sommets
    part = [k for k in range(N)]
    for s in range(N):
        for v in G[s]:
            reunir(part,s,v)
    return part
```
